## **Création du dataset final**

### **Merge entre les tables costomer et order**

```python
data1 = pd.merge(df3, df6,how='inner', on='order_id')
```

### **Ajout des autres tables**

```python
data2 = pd.merge(data1, df1, on= 'customer_id', how='left')
data3 = pd.merge(data2, df4, on= 'order_id', how='left')
data4 = pd.merge(data3, df5, on= 'order_id', how='left')
data5 = pd.merge(data4, df7, on= 'product_id', how='left')
data6 = pd.merge(data5, df8, on= 'seller_id', how='left')
data = pd.merge(data5, df9, on= 'product_category_name', how='left')
```

### **Répartition des clients par nombre de produist achetés**

```python
dft = df1.customer_unique_id.value_counts()
```

```python
plt.figure(figsize=(15, 6))
pourcentages = dft.value_counts(normalize=True) * 100
pourcentages.plot(kind='bar')
plt.ylabel('Pourcentage (%)')
plt.title('Répartition des clients par nombre de produist achetés')
for i, pourcentage in enumerate(pourcentages):
    plt.text(i, pourcentage + 1, f'{pourcentage:.1f}%', ha='center')

plt.show()
```

On remarque qu'effectivement seul 3% des clients ont fait 2 ou plusieurs commandes

### **Répresentation des 10 catégories de produist les plus achetés**

```python
data["product_category_name_english"].value_counts()[:10].plot(kind='bar')
plt.title('Répresentation des 10 catégories de produist les plus achetés');
```

Ce graphique affiche les 10 catégories de produits les plus achetées par les clients

### **Nombre de produist achetés par catégorie**

```python
plt.figure(figsize=(15,5))
dfx =data["product_category_name_english"].value_counts()
sns.barplot(data=dfx)
plt.xticks(rotation=90)
plt.title('Nombre de produist achetés par catégorie');
```

### **Répartion des paiement de produits par montant**

```python
plt.figure(figsize=(15,5))
dty =data["payment_value"].value_counts()
sns.histplot(data=dty, binwidth=5)
plt.title('Répartion des paiement de produits par montant')
plt.show();
```
